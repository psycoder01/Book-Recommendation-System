```python
import unittest
from ipynb.fs.full.Collab_filter import get_suggested_books

class FunctionalTest(unittest.TestCase):

    def test_movie_suggestions(self):
        test_book = "2nd Chance"
        self.assertIsNotNone(get_suggested_books(test_book))
        
    def test_random_input(self):
        test_book = "Some random string"
        self.assertEqual(get_suggested_books(test_book), [])
    
unittest.main(argv=[''], verbosity=2, exit=False)
```
