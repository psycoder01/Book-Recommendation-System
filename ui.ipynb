```python
import io
import requests
from tkinter import *
from PIL import Image, ImageTk
from ipynb.fs.full.Collab_filter import get_suggested_books, get_book

labels = []
buttons = []

def view_book(book_name):
    item_frame = Toplevel()
    item_frame.geometry("700x700")
    book = get_book(book_name)
    title = Label(item_frame , text = book_name)
    title.pack()
    author = Label(item_frame, text = "Author : " + book["author"])
    year = Label(item_frame, text = "Published Year : " + book["year"])
    publisher = Label(item_frame, text = "Published By : " + book["publisher"])
    isbn = Label(item_frame, text = "ISBN : " + book["ISBN"])
    image_view = Label(item_frame)
    
    author.pack()
    year.pack()
    publisher.pack()
    isbn.pack()
    image_view.pack()
    
    try:
        r = requests.get(book["image"])
        fp = io.BytesIO(r.content)
        image = Image.open(fp)
        photo = ImageTk.PhotoImage(image)
        image_view.config(image=photo)
        image_view.image = photo

    except requests.exceptions.RequestException as error:
        print(error)
    
def print_data(books):
    books_data = books
    del labels[:] # remove any previous labels from if the callback was called before
    del buttons[:]
    results_frame = Frame(root, width=400, height=300, bd=2, relief=GROOVE)
    results_frame.pack()
    i = 0
    for book in books_data:
        labels.append(Label( results_frame, text = book ))
        labels[i].grid( row = i , column = 0 )
        buttons.append(Button(results_frame,text="View", command = lambda name=book : view_book(name)))
        buttons[i].grid( row = i, column = 1)
        # Button(results_frame,text="View", command = lambda name=book : view_book(name)).grid( row = i , column = 1 )
        i = i + 1
    
def open_book(name):
    toplevel = Toplevel()
    label1 = Label(toplevel, text=name, height=0, width=100)
    label1.pack()

def get_input():
    books_list["text"] = ""
    error["text"]= ""
    input_book = entry1.get()
    
    if not input_book:
        error["text"] = "Please provide a book name"
        return "Please provide a book name"
        
    try:
        suggested_books = get_suggested_books(input_book)
         # If list is empty show error message
        if not suggested_books:
            error["text"] = "Please provide books name from given list for testing"
            return "Please provide books name from given list for testing"
    except:
        error["text"] = "Some error occured!"
        return "Some error occured!"
    
    print_data(suggested_books)

root = Tk()
root.geometry("1024x768")

#Label 1
label1 = Label(root,text = 'Enter a book name')
label1.pack()
label1.config(justify = CENTER)

entry1 = Entry(root, width = 30)
entry1.pack()

label2 = Label(root, text="For Test : Harry Potter and the Chamber of Secrets (Book 2) , 2nd Chance, Call of the Wild ")
label2.pack()
label1.config(justify = CENTER)

button1 = Button(root, text = 'Get suggestions')
button1.pack() 
button1.config(command = get_input)

label3 = Label(root, text="Suggested boooks : ")
label3.pack()
label1.config(justify = CENTER)
books_list = Label(root, text= "" )
books_list.pack()
books_list.config(justify = CENTER)
    
error = Label(root, text="")
error.pack()
label1.config(justify = CENTER)

root.mainloop()
```
